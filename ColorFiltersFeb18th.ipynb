```python
!pip install opencv-python
!pip install requests
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt
import cv2

```

```python
url = 'https://media.globalcitizen.org/thumbnails/97/76/97766969-a665-4fe5-871f-b01abc7bc7e8/world-giraffe-day-hero-2.jpg__1600x900_q85_crop_subsampling-2.jpg'
image = np.asarray(Image.open(requests.get(url, stream=True).raw))

def rgb_splitter(image):
    rgb_list = ['Reds', 'Greens', 'Blues']
    fig, ax = plt.subplots(1, 7, figsize=(20, 5))
    rgb_filter = []
    ax[0].imshow(image)
    ax[0].set_title('Original', fontsize=15)
    ax[0].axis('off')

    for i in range(3):
        ax[i+1].imshow(image[:, :, i], cmap=rgb_list[i])
        ax[i+1].set_title(rgb_list[i], fontsize=15)
        ax[i+1].axis('off')

    def remove_blue(image):
        filtered_photo = np.copy(image)
        filtered_photo[:, :, 2] = 0
        return filtered_photo, "Red and Green"

    def remove_green(image):
        filtered_photo = np.copy(image)
        filtered_photo[:, :, 1] = 0
        return filtered_photo, "Red and Blue"

    def remove_red(image):
        filtered_photo = np.copy(image)
        filtered_photo[:, :, 0] = 0
        return filtered_photo, "Green and Blue"

    filtered_images = [remove_blue(image), remove_green(image), remove_red(image)]
    for i in range(0, 3):
      ax[i+4].imshow(filtered_images[i][0])
      ax[i+4].set_title(filtered_images[i][1], fontsize=15)
      ax[i+4].axis('off')

    plt.show()

rgb_splitter(image)

red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

combined_image = cv2.merge([red, green, blue])

if np.array_equal(combined_image, image):
  print("Original Image and Image are identical")

else:
  print("Try again!")
```
