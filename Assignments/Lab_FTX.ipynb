## 1. Load Libraries & Data

Import any necessary libraries and load the remote CSV file below into a DataFrame. **Resolve any import issues you need to.**

```python
import pandas as pd
import numpy as np
column_names = 'timeOpen', 'timeClose',	'timeHigh',	'timeLow',	'open',	'high',	'low',	'close',	'volume', 'marketCap',	'timestamp'
df = pd.read_csv('https://s3.amazonaws.com/ds1002-resources/data/ftx-2021-01-01--2023-12-31.csv', delimiter=';', header=None, names=column_names)
```

## 2. Get Summary Data

In code, display how many rows and columns are in the raw dataset.

```python
#Drops the first row because it was reading the first row as the columns
df = df.iloc[1:]
df
```

```python
num_rows, num_columns = df.shape
print('The data set has', num_rows, 'rows')
print('The data set has', num_columns, 'columns')
```

## 3. Clean and Organize the Data

Check for data quality.

- Resolve any empty values or duplicate rows.
- Order the data properly if necessary.
- Drop the following columns and save the dataframe: `timeOpen`, `timeClose`, `timeHigh`, `timeLow`, `marketCap`



```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
columnsdrop = ['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'marketCap']
df = df.drop(columnsdrop, axis=1)
df
```

## 4. Calculate Daily Gains/Losses

Using Pandas, add a new column named `["close_change"]` to the dataframe that calculates the difference between daily closing prices for FTX. This will result in positive (growth) and negative (loss) values.

Note:
- Be careful about how you calculate this. Each row should calculate that difference from the row above it, which must be the previous calendar day.
- Hint: This is not complicated and does not require a for-loop.  See the Pandas review notebook for reference.

```python
df['close'] = df['close'].astype(float)
df['close_change'] = df['close'] - df['close'].shift(1)
#Filling the first row in close_change to 0, because it would otherwise come back as NaN because it has no value to compare to)
df = df.fillna('0')
df.head()
```

## 5. Identify Growth Days

Based on the daily change in closing prices calculated above, identify the top 10 days of positive growth for FTX.

Display your results in a table showing both the rate of change and their dates.

```python
#It kept giving me error messages for floats so I converted the object type in pretty much every chunk
df['close_change'] = df['close_change'].astype(float)
#nlargest finds n largest numbers, 10 is the given number
top_pos_growth = df.nlargest(10, 'close_change')
print(top_pos_growth[['timestamp', 'close_change']])
```

## 6. Calculate the Percentage Rate of Change

Next, create a new column named `["rate_of_change"]` based on `["close_change"]`.

Use the Pandas Review notebook for reference. Again, this is an easy operation in Pandas and does not require a for-loop.

```python
df['rate_of_change'] = df.apply(lambda row: (row['close_change'] / row['close']) * 100, axis=1)
df
```

## 7. Identify Days of Greatest Percentage Change

Based on the rate change column calculated above, identify the top 5 days of greatest positive rate change, and the top 5 days of greatest negative rate change.

Display your results in a table showing both the rate of change and their dates.

```python
print('Top 5 days of greatest positive rate change')
#nlargest finds n largest numbers, 5 is the given number
top_pos_rate_change = df.nlargest(5, 'rate_of_change')
print(top_pos_rate_change[['timestamp', 'rate_of_change']])
```

```python
print('Top 5 days of greatest negative rate change')
#nsmallest finds n smallest numbers, 5 is the given number
top_neg_rate_change = df.nsmallest(5, 'rate_of_change')
print(top_neg_rate_change[['timestamp', 'rate_of_change']])
```

## 8. Plot the Data

Create two separate plots of the data -- one for the daily closing price and one for the daily gains/losses. These can be histograms or line plots.

```python
import matplotlib.pyplot as plt
#convert to floats because there was a type error
df['open'] = df['open'].astype(float)
df['close'] = df['close'].astype(float)
df['daily_numbers'] = df['open'] - df['close']
dailygainsorlosses = plt.hist(df['daily_numbers'])
plt.title("Daily Gains/Losses")
plt.xlabel("Frequency")
plt.ylabel("Gains/Losses")
plt.show()
```

```python
closingpricegraph = plt.hist(df['close'])
plt.title("Daily Closing Price")
plt.xlabel("Frequency")
plt.ylabel("Closing Price")
plt.show()
```

Combine the two plots from above into a single plot, displaying both together. Your plot should have two Y-axis scales, should differentiate the two values by color, and should include a legend.

```python
fig, ax1 = plt.subplots()
#Alpha allows for both plots to be semi-transparent allowing to see the overlays
ax1.hist(df['daily_numbers'], color='pink', alpha=0.7)
ax1.set_xlabel('Frequency')
ax1.set_ylabel('Gains/Losses', color='pink')
ax1.tick_params(axis='y', labelcolor='pink')

#twinx allows there to be two y axis with one x axis
ax2 = ax1.twinx()

ax2.hist(df['close'], color='blue', alpha=0.7)
ax2.set_ylabel('Closing Price', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')


plt.title('Combined Plot of Daily Gains/Losses and Closing Price')


ax1.legend(['Gains/Losses'], loc='upper left')
ax2.legend(['Closing Price'], loc='upper right')

plt.show()
```
